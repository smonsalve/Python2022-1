Exercise 104: Sorted Order

(Solved—21 Lines) Write a program that reads integers from the user and stores them in a list. Your program should continue reading values until the user enters 0. Then it should display all of the values entered by the user (except for the 0) in order from smallest to largest, with one value appearing on each line. Use either the sort method or the sorted
function to sort the list.

```python
milista = []
valor = int(input("Por favor ingrese un entero: "))

while(valor != 0):
  milista.append(valor)
  valor = int(input("Por favor ingrese un entero: "))

milista = sorted(milista)

print(milista)
```
