```python
psw = input("ingrese su contraseña: ")

def validador(psw):
    conta= {"mayu": 0,"minu": 0, "num":0}

    for caracter in psw:
        if caracter.isupper():
            conta["mayu"] += 1

        elif caracter.islower():
            conta["minu"] += 1
        
        if caracter.isnumeric():
            conta["num"] += 1

    if conta["num"] > 0:
        print("validador de numeros correcto.")
    else:
        print("la clave require al menos un numero.")
    if conta["mayu"] > 0 and conta["minu"] > 0:
        print("validación de mayus y minus correcta.")
    elif conta["mayu"] > 0 and conta["minu"] == 0:
        print("la contraseña requiere al menos una minuscula.")
    elif conta["mayu"] == 0 and conta["minu"] > 0:
        print("la contraseña necesital menos una mayuscula.")
    elif conta["mayu"] == 0 and conta["minu"] == 0:
        print("la contraseña necesitas al menos una mayuscula y una minuscula.")

    if len(psw) == 8:
        print("contraseña tiene 8 caracteres.")
    else:
        print("la contraseña debe tener 8 caracteres, intenta de nuevo.")   
validador(psw)
```
