An integer, n, is said to be perfect when the sum of all of the proper divisors of n is equal to n. 
For example, 28 is a perfect number because its proper divisors are 
1, 2, 4, 7 and 14, and 1 + 2 + 4 + 7 + 14 = 28.
Write a function that determines whether or not 
a positive integer is perfect. 
Your function will take one parameter. 
If that parameter is a perfect number 
then your function will return true. 
Otherwise it will return false. 
In addition, write a main program 
that uses your function to identify and display 
all of the perfect numbers between 1 and 10,000. 
Import your solution to Exercise 109 when completing this task.

```python
def perfecto(x):
  suma = 0
  divisores = []
  for i in range(1,x):
    if(x%i==0):
      divisores.append(i)
      suma += i
  return(True if (suma==x) else False)  
for i in range(1,10000):
  if(perfecto(i)): print(i)
```

```python
def perfect(x):
  divisores_propios = []
  suma = 0
  for i in range(1,x):
    if x%i==0:
      divisores_propios.append(i)
  
  for j in divisores_propios:
    suma += j

  if(suma == x): 
    return True
  else:
    return False

for i in range(29):
  print(i,perfect(i))
```
